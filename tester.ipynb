```python
import mpyq

replay_path = "data/Blizzard/set_1/00a0a1b139395dbbba8058a0ec42128b2356cf92abd1dd0ae7059692c37124be.SC2Replay"
archive = mpyq.MPQArchive(replay_path)
```

```python
print(archive.files)
```

```python
contents = archive.header['user_data_header']['content']
from s2protocol import versions
header = versions.latest().decode_replay_header(contents)
baseBuild = header['m_version']['m_baseBuild']
protocol = versions.build(baseBuild)
```

```python
contents = archive.read_file('replay.game.events')
gameEvents = protocol.decode_replay_game_events(contents)
events_list = []
for event in gameEvents:
    events_list.append(event)
    
```

```python
import json 
json_data = json.dumps(events_list, indent=4)
output_file = open('C:/Users/jared/Desktop/output.json', 'w')
output_file.write(json_data)
output_file.close()
```
